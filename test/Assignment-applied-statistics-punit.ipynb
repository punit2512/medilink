```python
# Checking outliers for age
mean_age, std_age = np.mean(insurance.age), np.std(insurance.age)
cut_off = std_age * 3
lower_age, upper_age = mean_age - cut_off, mean_age + cut_off
outliers_age = [x for x in insurance.age if x < lower_age or x > upper_age]
outliers_age
```

```python
# No outliers in age identified
```

```python
#Checking outliers for BMI
mean_bmi, std_bmi = np.mean(insurance.bmi), np.std(insurance.bmi)
cut_off = std_bmi * 3
lower_bmi, upper_bmi = mean_bmi - cut_off, mean_bmi + cut_off
outliers_bmi = [x for x in insurance.bmi if x < lower_bmi or x > upper_bmi]
outliers_bmi
```

```python
#There are some outliers in bmi, as also can be seen from the scatter plot.
```

```python
#Checking outliers for charges
mean_charges, std_charges = np.mean(insurance.charges), np.std(insurance.charges)
cut_off = std_charges * 3
lower_charges, upper_charges = mean_charges - cut_off, mean_charges + cut_off
outliers_charges = [x for x in insurance.charges if x < lower_charges or x > upper_charges]
outliers_charges
```

```python
#This shows we have outliers for charges too which are also shown in the scatter plot.
```
